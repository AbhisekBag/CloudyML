### Will see `How to make your member variables private?`
* A double underscore makes variable Private which makes it harder to access but still possible
* Variable with one leading underscore is semi-private

Using Encapsulation in Python

```python
class car:
        def __init__(self, speed, color):
                self.speed = speed
                self.color = color
                print('init is called')

ford = car(speed=45, color='blk')

print(ford.speed)
ford.speed = 'changed'
print(ford.speed)
```

```python
class car:
        def __init__(self, speed, color):
                self.speed = speed
                self.color = color
        def set_speed(self, value):
                self.speed = value

        def get_speed(self):
                return self.speed
        

ford = car(speed=45, color='blk')

print(ford.speed)
ford.set_speed(200)
print(ford.get_speed())
```

If anyone came and gave some string values to ford.speed which will lead to code break and maybe some loss, so we require `Encapsulation`
* it is very important to protect our data and only give acess to required data to other user
* it is very important when you want ot give your code to other people and they want to change it

**`To Encapsulate code we use Functions`**

```python
class car:
        def __init__(self, speed, color):
                self.speed = speed
                self.color = color
        def set_speed(self, value):
                self.speed = value

        def get_speed(self):
                return self.speed
        

ford = car(speed=45, color='blk')

print(ford.speed)
ford.set_speed(200)
ford.speed = 400
print(ford.get_speed())
```

<u><center> <h3 style="background-color:orange"><br>Inheritance<br></h3></center></u>


Classes in Python can be extended
* creating new classes which retain characteristics of base class and this process is known as `Inheritance`

![](https://www.codesdope.com/pa-images-bucket/courses/python/sub1.jpg)

Python Inheritance Involves two kind of classes a `Superclass`  and a `Subclass`
* Subclass inherits members of Superclass, on top of which it can add its own members.

Idea of inheritance is that a class can be defined to borrow behavior (member variable and member methods) from another class.
* When we define a subclass we place name of parent(superclass) in parentheses after subclassâ€™s(child) name on first line of definition
   
   '''class Parent(Child):'''
  
See Superclass and Subclass as Mobile(Superclass) and iphone,lava,samsung(Subclass)
* Relationship between Superclass and Subclass is shown by **is a** relationship
    * iphone isa Mobile
  
   

```python
class mobile:
    __imei = None
    __model = None
    def set_values(self, imei, model):
        self.__imei = imei
        self.__model = model

class iphone(mobile):
    def iphone_detail(self):
        return self.__imei, self.__model
    
iphone = iphone()
iphone.set_values('123456789', 'iphone 11')

print(iphone.iphone_detail())
```

`ERROR:` as `__imei and __model` are Private member variables of Superclass i.e. Mobile class and cannot be accessed by Subclass

Point is ==> Private member of Superclass can't be inharited by Subclass

**Solution is to make a getter method for private members**

```python
class mobile:
    __imei = None
    __model = None
    def set_values(self, imei, model):
        self.__imei = imei
        self.__model = model
    def get_imei(self):
        return self.__imei
    def get_model(self):
        return self.__model

class iphone(mobile):
    def iphone_detail(self):
        return self.get_imei(), self.get_model()
    
iphone = iphone()
iphone.set_values('123456789', 'iphone 11')

print(iphone.iphone_detail())
```
