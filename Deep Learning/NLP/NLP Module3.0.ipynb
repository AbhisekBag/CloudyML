 **So further we are go to Take Paragraph as input variable and we will Tokenize the paragraph into sentences and words respectively.**



```python
#Run this cell for storing input paragraph into one variable
paragraph = '''Welcome to CloudyML. Hands-on of every thing you need to know about Cloud & Machine Learning and how we can combine both to build very powerful applications corpora.'''
```

In the next cell,<br>
the paragraph is converted into sentences and  sent_tokenize method from the nltk library is used to tokenize sentences

```python
# Tokenizing words using nltk.word_tokenize
#(converting paragraphs into words) 
all_words = [nltk.tokenize.word_tokenize(tokens) for tokens in tokens]
all_words
```

# **Lemmatization**

It is the process of grouping together the different inflected forms of a word so they can be analyzed as a single item.
It is similar to stemming but it brings context to the words. So it links words with similar meanings to one word.
2_lemma.png


**Applications of lemmatization are:**

Used in comprehensive retrieval systems like search engines and compact indexing

# **Stemming**

It is the process of producing morphological variants of a root/base word. Stemming programs are commonly referred to as stemming algorithms or stemmers. A stemming algorithm reduces the words chocolates, chocolatey, choco to the root word, chocolate and retrieval, retrieved, retrieves reduce to the stem retrieve. Stemming is an important part of the pipelining process in Natural language processing. The input to the stemmer is tokenized words.

4_Stemming.jpeg

Applications of stemming : 
1.  Stemming is used in information retrieval systems like search engines.
2.  It is used to determine domain vocabularies in domain analysis.




---
Same as above we are going importing nltk and porterstemmer for stemming 

---



The Porter stemming algorithm (or 'Porter stemmer') is a process for removing the commoner morphological and inflexional endings from words in English. Its main use is as part of a term normalisation process that is usually done when setting up Information Retrieval systems.

Reference: https://www.nltk.org/_modules/nltk/stem/porter.html

```python
# import these modules 
# from nltk.stem import porterstemmer
from nltk.stem import PorterStemmer
```


In the next cell,<br>
ps(any name can be chosen) object of the PorterStemmer class is created as object is an instance of the class that can implement all the methods and functionalities of the class.

```python
# Create PorterStemmer Object
stem  = PorterStemmer()
```


Stem method of the PorterStemmer class is called by the 'ps' object to stem the given words.

```python
#choose some words to be stemmed
words = ["program", "programs", "programmer", "programming", "programmers"]

#use for loop to iterate words
for word in words:
    print(stem.stem(word))

# stem method of the PorterStemmer class is called by the 'ps' object to stem the given words.
```


As we can that the various words of programs is stemmed into program.

##Stemming VS Lemmatization

**Stemming** just removes or stems the last few characters of a word, often leading to incorrect meanings and spelling.<br>
<br>
**Lemmatization** considers the context and converts the word to its meaningful base form, which is called Lemma. Sometimes, the same word can have multiple different Lemmas.

# How to remove stop words
The stopwords in nltk are the most common words in data. They are words that you do not want to use to describe the topic of your content. They are pre-defined and cannot be removed.

![image24.png](https://www.askpython.com/wp-content/uploads/2020/11/stop-words-1024x512.png.webp)

Documtation: https://pythonspot.com/nltk-stop-words/

```python
#import nltk and download nltk stopwords
#nltk.download('stopwords')
nltk.download('stopwords')
```

---
Further, English stopwords are loaded using the 'words' method from the stopwords class.

---


```python
#importing Stop words
from nltk.corpus import stopwords 
#stopwords.words('english')
```



In the next Cell,
english stopwords are loaded into the *en_stops* variable using the set function

```python
#Checking Stops words in a sentences 
#example of Stopwords 
#you can try with your inputs :)
en_stops = set(stopwords.words('english'))
words = ['There', 'is', 'a', 'Deep','Learning','Course','in','CLoudyML']
for word in words: 
    if word not in en_stops:
        print(word)
```


So as we can see that 'is', 'a' , 'in' is removed because they are stopswords.
